```python
import pandas as pd
import os
from SPARQLWrapper import SPARQLWrapper, JSON
from string import Template
```

```python
#This program reads a csv file that has a column of CAS numbers and returns the equivalent NVS S27 code
pf = pd.read_csv("casNO.csv")
pf.head()
print("Starting here")#Just to start
vs=(pf.size)
sparql = SPARQLWrapper("http://vocab.nerc.ac.uk/sparql/sparql") #Choose the sparql endpoint you wish
query=("""
prefix skos:<http://www.w3.org/2004/02/skos/core#>
prefix rdf:<http://www.w3.org/1999/02/22-rdf-syntax-ns#>
prefix owl:<http://www.w3.org/2002/07/owl#>
prefix dc:<http://purl.org/dc/terms/>
PREFIX injection: <http://www.morelab.deusto.es/injection.owl#>

select distinct (?k as ?uri) (?kdef as ?definition) (?kpl as ?prefLabel) (?kaltpl as ?altLabel)  where
{<http://vocab.nerc.ac.uk/collection/P01/current/> skos:member ?dt .  ?dt skos:prefLabel ?pl ;
skos:broader ?k . FILTER (regex(str(?k),"S27","i") )
?k skos:definition ?kdef .
?k skos:prefLabel ?kpl .
?k skos:altLabel ?kaltpl .
FILTER (regex(str(?pl),'CASNo',"i") )
}
""")

for index, row in pf.iterrows():
    s=row[0]
    q=query.replace("CASNo",s)
    sparql.setQuery(q)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    for result in results["results"]["bindings"]:
       print(s+"   "+result["uri"]["value"],result["definition"]["value"],result["prefLabel"]["value"],result["altLabel"]["value"],sep="---")
```
